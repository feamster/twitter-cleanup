```python
from time import sleep, strftime, time

TIME = time()

def sleepy_time(mode, sw):
    # sw: friends or messages
    # return: float cooldown seconds
    global TIME

    t = strftime('%H:%M:%S')
    if sw == 'friends':
        lim = 0  # correct
    elif sw == 'messages':
        lim = 180 if mode == "user" else 60
    elif sw == 'timeline':
        lim = 180 if mode == "user" else 300
    else:
    	lim = 15 if mode == "user" else 30

    process_time = time() - TIME
    TIME = time()
    cooldown = float(15 * 60) / float(lim) - process_time
    print "sleeping", cooldown

    return cooldown if cooldown > 0 else 0
```
